```python
#option 1: use the datetime stamp
from datetime import datetime
```

```python
mar=datetime(2021, 3, 17)
stamp_mar = datetime.timestamp(mar)
print(stamp_mar)
apr=datetime(2021,4,17)
stamp_apr = datetime.timestamp(apr)
print(stamp_apr)
# datetime(year, month, day[, hour[, minute[, second[, microsecond[,tzinfo]]]]
```

```python
comments_month = SITE.fetch('comments', fromdate=int(stamp_mar), todate=int(stamp_apr))
comments_month = pd.DataFrame(comments_month)
comments_month.head()
```

```python
#OPTION2: use "commonly" formatted datetime
comments_month2 = SITE.fetch('comments', fromdate=(2021,3,17), todate=(2021,4,17))
comments_month2 = pd.DataFrame(comments_month)
comments_month2.head()
```

Not part of the assignment, but just to practice a little: flatten some columns and append them to the dataframe.

```python
#For my own sake, lets try to flatten the nested column 'items'.

items_dict = dict(df_comments['items']) #make a dict out of the nested column
items_flat = pd.DataFrame(items_dict).transpose() #put the dict in a dataframe
columns = [str(i) for i in items_flat.columns] #make strings of the column names
items_flat.columns= ['items' + '_' + i for i in columns] #change the column names (add prefix)
items_flat.columns #inspect the new column names
```

```python
items_flat.head() #inspect the dataframe with the added unnested columns
```

```python
#further unnesting of 'items_owner'- column and appending it to the items_flat.
items_owner_dict = dict(items_flat['items_owner']) #making a dict out of the column
items_owner_flat = pd.DataFrame(items_owner_dict).transpose() #put the dict into a dataframe
columns = [str(i) for i in items_owner_flat.columns] #make a string out of the table headers
items_owner_flat.columns= ['items_owner' + '_' + i for i in columns] #add prefix to the name
items_owner_flat.columns #check the col names
```

```python
items_owner_flat.head()
```

```python
items_flat = pd.concat([items_flat, items_owner_flat], axis = 1) #add this unpacked items_owner to items_flat
items_flat.head()
```

```python
items_flat = items_flat.drop('items_owner', axis = 1) #drop the items_owner column as you've unpacked this already
```

```python
# There is more unnesting to be done in items_reply_to_user, let's just do it.

reply = dict(items_flat["items_reply_to_user"]) #take the nested column and turn in into a dict:
reply_flat = pd.DataFrame(reply).transpose() #put the dict in a dataframe
reply_flat.head()
columns= [str(i) for i in reply_flat.columns] #make string out of column-names
reply_flat.columns= ['reply_items' + '_' + i for i in columns] #add where you got it from, in this case: reply_items_[name]
reply_flat.head()
```

```python
items_flat = pd.concat([items_flat, reply_flat], axis = 1) #append this to items_flat.
items_flat.head()
```

```python
#now concat to df_comments, lets name is df_comments2

df_comments2 = pd.concat([df_comments, items_flat], axis=1)
df_comments2.head()
```

```python
#drop the items en items_owner columns
df_comments2 = df_comments2.drop(['items', 'items_owner'], axis = 1)
```
